```python
from scipy.spatial.distance import cosine
from scipy.stats import pearsonr
from math import *
import numpy as np
```

```python
# 피어슨 상관계수
def pearson_similarity(x, y):
    return np.dot((a - np.mean(a)), (b - np.mean(b))) / ((np.linalg.norm(a - np.mean(a))) * (np.linalg.norm(b - np.mean(b))))
```

```python
# 코사인 유사도
def cosine_similarity(a, b):
    return np.dot(a, b.T) / (np.linalg.norm(a) * (np.linalg.norm(b)))
```

```python
# 자카드 유사도 - X(집합에 이용)
```

```python
# 유클리드 유사도
def euclidean_distance(a, b):
    distance = 0
    for in in range(len(a)):
        distance += (a[i] - b[i]) ** 2
    return distance ** 0.5


def Euclidean(a, b):
    vec1 = a.copy()
    vec2 = b.copy()
    if len(vec1)<len(vec2): vec1,vec2 = vec2,vec1
    vec2 = np.resize(vec2,(vec1.shape[0],vec1.shape[1]))
    return np.linalg.norm(vec1-vec2)
```

```python
# 삼각형 면적 유사도 (TS 유사도)
def Theta(a, b):
    return np.arccos(Cosine(a, b)) + np.radians(10)
    
def triangle_area_similarity(a, b):
    theta = np.radians(Theta(a, b))
    return ((np.linalg.norm(a) * np.linalg.norm(b)) * np.sin(theta))/2

# 부채꼴 면적 유사도 (SS 유사도)
def Magnitude_Difference(vec1, vec2):
        return abs((np.linalg.norm(vec1) - np.linalg.norm(vec2)))

def sector_area_similarity(a, b):
    ED = Euclidean(a, b)
    MD = Magnitude_Difference(a, b)
    theta = Theta(a, b)
    return math.pi * (ED + MD)**2 * theta/360

## TS-SS 유사도
def ts_ss_similarity(a, b):
    return triangle_area_similarity(a,b) * sector_area_similarity(a,b)
```

```python
# 조정된 코사인 유사도 (두 개의 사용자 아이템)
```

```python
def recommendation():
    user = normalization_df.loc['user'].values
    for idx in u_df.index:
        item = normalization_df.loc[idx].values
        # 코사인 유사도
        cosine = 1 - cosine(item, user)
        result.loc[idx, '코사인'] = cosine
        # 피어슨 상관계수
        pearson = pearsonr(item, user)
        result.loc[idx, '피어슨'] = pearson
        # 유클리드 유사도
        euclidean = euclidean_distance(item, user)
        result.loc[idx, '유클리드'] = euclidean
        # TS 유사도
        triangle = triangle_area_similarity(item, user)
        result.loc[idx, 'TS'] = triangle
        # SS 유사도
        sector = sector_area_similarity(item, user)
        result.loc[idx, 'SS'] = sector
        # TS-SS 유사도
        ts_ss = ts_ss_similarity(item, user)
        result.loc[idx, 'TS-SS'] = ts_ss
    return result
    
```

```python
recommendation()
```

```python
result = result.drop(['user'], axis = 0)
```
